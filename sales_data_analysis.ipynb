```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ECOMM_DATA.csv', encoding='latin-1')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
#Data Exploration & Cleaning

df.isnull().sum()
```

```python
df.drop_duplicates(inplace = True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.Sales.value_counts()
```

```python
# Calculate total sales
total_sales = df['Sales'].sum()
print("Total Sales: ${:,.2f}".format(total_sales))
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.month
monthly_sales = df.groupby('Month')['Sales'].sum()
print(monthly_sales)
```

```python
monthly_sales.replace([float('inf'), float('-inf')], float('nan'), inplace=True)
```

```python
#Monthly Sales Trends

plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_sales, marker='o', color='b')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
# Group data by country and calculate total sales
total_sales_by_country = df.groupby('Country')['Sales'].sum()

# Sort by total sales in descending order (highest to lowest)
top_10_countries = total_sales_by_country.sort_values(ascending=False).head(10)

print("Top 10 Countries by Sales:")
print(top_10_countries)
```

```python
#Top 10 countries by sales
total_sales_by_country = df.groupby('Country')['Sales'].sum()

top_10_countries = total_sales_by_country.sort_values(ascending=False).head(10)

# bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_10_countries.index, top_10_countries.values, color='skyblue')
plt.xlabel("Country")
plt.ylabel("Total Sales")
plt.title("Top 10 Countries by Sales")
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

```python
# Best-Selling Products
best_selling_products = df.groupby('Product Name')['Quantity'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Best-Selling Products:")
print(best_selling_products)
```

```python
plt.figure(figsize=(12, 8))
best_selling_products.plot(kind='bar', color='skyblue')
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Sort products by total sales in descending order
top_selling_products = best_selling_products.sort_values(ascending=False)

# Bar chart for top-selling products
plt.figure(figsize=(12, 6))
plt.bar(top_selling_products.index[:10], top_selling_products.values[:10], color='goldenrod')  # Consider limiting to top 10
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.title("Top-Selling Products")
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()

#Pie chart for market share of products
plt.figure(figsize=(8, 8))
plt.pie(top_selling_products.head(10), labels=top_selling_products.head(10).index, autopct="%1.1f%%", startangle=140)
plt.title("Market Share of Top Products")
plt.axis("equal") 
plt.tight_layout()
plt.show()
```

```python

# Calculate total sales per product
total_sales_by_product = df.groupby('Product Name')['Sales'].sum()

# Sort products by total sales in descending order
top_selling_products = total_sales_by_product.sort_values(ascending=False)

#Bar chart for top-selling products
plt.figure(figsize=(12, 6))
plt.bar(top_selling_products.index[:10], top_selling_products.values[:10], color='goldenrod') 
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.title("Top-Selling Products")
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()

# Pie chart for market share of products
plt.figure(figsize=(8, 8))
plt.pie(top_selling_products.head(10), labels=top_selling_products.head(10).index, autopct="%1.1f%%", startangle=140)
plt.title("Market Share of Top Products")
plt.axis("equal") 
plt.tight_layout()
plt.show()
```

```python
#Profit Analysis by Segment
plt.figure(figsize=(10, 6))
sns.boxplot(x='Segment', y='Profit', data=df, palette='Set2')
plt.title('Profit Distribution by Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Profit')
plt.xticks(rotation=45) 
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
profit_by_segment = df.groupby('Segment')['Profit'].sum()

plt.figure(figsize=(8, 8))
plt.pie(profit_by_segment, labels=profit_by_segment.index, autopct="%1.1f%%", startangle=140)
plt.title("Profit Distribution by Segment")
plt.axis("equal") 
plt.tight_layout()
plt.show()
```

```python

#Order Priority Analysis using a Count Plot
plt.figure(figsize=(8, 6))
sns.countplot(x='Order Priority', data=df, palette='viridis')
plt.title('Distribution of Order Priorities')
plt.xlabel('Order Priority')
plt.ylabel('Count')
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

```python

#Calculate the average number of orders per customer
average_orders_per_customer = df.groupby('Customer ID')['Order ID'].count().mean()

# Visualize the distribution of the average number of orders
plt.figure(figsize=(8, 6))
sns.histplot(df.groupby('Customer ID')['Order ID'].count(), bins=20, kde=True, color='skyblue')
plt.axvline(average_orders_per_customer, color='red', linestyle='--', label='Average Orders per Customer')
plt.title('Distribution of Average Orders per Customer')
plt.xlabel('Number of Orders')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Fulfillment Time'] = (df['Ship Date'] - df['Order Date']).dt.days

# Calculate the average time taken to fulfill orders
average_fulfillment_time = df['Fulfillment Time'].mean()

#Visualize the distribution of the time taken to fulfill orders
plt.figure(figsize=(8, 6))
sns.histplot(df['Fulfillment Time'], bins=20, kde=True, color='skyblue')
plt.axvline(average_fulfillment_time, color='red', linestyle='--', label='Average Fulfillment Time')
plt.title('Distribution of Order Fulfillment Time')
plt.xlabel('Fulfillment Time (days)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
