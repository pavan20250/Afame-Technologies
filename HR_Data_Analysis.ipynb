```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('HR Data.csv', encoding='latin-1')
df.head()
```

```python
# Remove unnecessary columns

columns_to_drop = ['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours']
df.drop(columns=columns_to_drop, inplace=True)
print(df.head())
```

```python
# Rename columns
# In this example, let's rename the 'ï»¿Age' column to 'Age'

df.rename(columns={'ï»¿Age': 'Age'}, inplace=True)

print(df.head())
```

```python
# Drop redundant entries 
df.drop_duplicates(inplace=True)

print(df.columns)
print(df.head())
```

```python
# Sanitize specific columns
df['MaritalStatus'] = df['MaritalStatus'].str.lower()

print(df['MaritalStatus'].unique())
```

```python
# Eliminate NaN values
df.dropna(inplace=True)

print(df.head())
```
